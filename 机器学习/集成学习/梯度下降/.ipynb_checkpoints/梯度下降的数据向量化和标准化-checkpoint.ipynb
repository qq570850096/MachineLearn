# 梯度下降的数据向量化和标准化
$$
\nabla{J}(\theta)=\frac{2}{m}·\begin{pmatrix}
\sum_{i=1}^{m}{(y^i-X_b^i\theta)(-1)}\\
\sum_{i=1}^{m}{(y^i-X_b^i\theta)(-X_1^i)}\\
\sum_{i=1}^{m}{(y^i-X_b^i\theta)(-X_2^i)}\\
……\\
\sum_{i=1}^{m}{(y^i-X_b^i\theta)(-X_n^i)}
\end{pmatrix}
=\frac{2}{m}·\begin{pmatrix}
\sum_{i=1}^{m}{(X_b^i\theta-y^i)(X_0^i)}\\
\sum_{i=1}^{m}{(X_b^i\theta-y^i)(X_1^i)}\\
\sum_{i=1}^{m}{(X_b^i\theta-y^i)(X_2^i)}\\
……\\
\sum_{i=1}^{m}{(X_b^i\theta-y^i)(X_n^i)}
\end{pmatrix}
$$

一个1\*m的矩阵和m*（n+1）的矩阵

所以原式可以化简为
$$\frac{2}{m}\cdot{(X_b\theta-y)^T}\cdot{X_b}=\frac{2}{m}\cdot{X_b^T}\cdot{(X_b\theta-y)}$$

```python
import numpy as np
from sklearn import datasets
```

```python
boston = datasets.load_boston()
x = boston.data
y = boston.target

x=x[y<50]
y=y[y<50]
```

```python
lin_reg = LinearRegression()
lin_reg.fit_normal(x_train,y_train)
```

## 梯度下降法的优势

```python
m = 1000
n = 5000

big_x = np.random.normal(size=(m,n))

true_theta = np.random.uniform(0.0,100.0,size=n+1)

big_y = big_x.dot(true_theta[1:])+true_theta[0]+np.random.normal(0.0,10.0,size=m)
```

```python
big_reg1 = LinearRegression()
%time big_reg1.fit_normal(big_x,big_y)
big_reg1.coef_
```

```python
big_reg2 = LinearRegression()
%time big_reg2.fit_gd(big_x,big_y)
big_reg2.coef_
```
