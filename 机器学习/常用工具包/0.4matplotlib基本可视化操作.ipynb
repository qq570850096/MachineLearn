```python
# -*- coding: utf-8 -*-
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
x
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y)
```

```python
cosy = np.cos(x)
cosy.shape
```

```python
siny = y.copy()
```

```python
plt.plot(x,siny,linestyle = '--')
plt.plot(x,cosy,color = '#6DC971')
```

```python
plt.plot(x,siny,linestyle = '--')
plt.plot(x,cosy,color = '#6DC971')
plt.ylim(0,1)
```

```python
plt.plot(x,siny,linestyle = '--')
plt.plot(x,cosy,color = '#6DC971')
plt.axis([-1,11,-2,2])
```

```python
plt.plot(x,siny,linestyle = '--',label = "sin(x)")
plt.plot(x,cosy,color = '#6DC971',label = "cos(x)")
plt.axis([-1,11,-2,2])
plt.xlabel("x value")
plt.ylabel("y value")
plt.legend()
plt.title("hello world")
```

```python
plt.scatter(x,siny,label='sin(x)')
plt.scatter(x,cosy,color='red',label='cos(x)')
plt.legend()
```
