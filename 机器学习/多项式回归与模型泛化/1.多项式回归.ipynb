## 多项式回归

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-3,3,100)
X = x.reshape(-1, 1)
y = 0.5*x**2 + x + 2 + np.random.normal(0,1,size = 100)
y.shape
```

```python
plt.scatter(X,y)
```

```python
import sys
sys.path.append('../')
```

```python
from playMachine.LinearRegression import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit_normal(X,y)
```

```python
plt.scatter(X,lin_reg.predict(X))
plt.scatter(X,y)
```

### 解决方案——添加一个特征

```python
X2 = np.hstack([X,X**2])
X2.shape
```

```python
lin_reg2 = LinearRegression()
lin_reg2.fit_normal(X2,y)
plt.scatter(X,lin_reg2.predict(X2))
plt.scatter(X,y)
```
