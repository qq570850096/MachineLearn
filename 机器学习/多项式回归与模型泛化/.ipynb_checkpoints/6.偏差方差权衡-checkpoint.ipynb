## 模型误差 = 偏差+方差+不可避免的误差

导致产生偏差的原因：对模型的假设不准确，比如对非线性数据使用了线性回归。

高偏差一般会导致欠拟合（underfitting）

高方差的原因：模型太过复杂，一点扰动都会极大地干扰模型，泛华能力差。

高方差一般会导致过拟合（overfitting）

## 有一些算法天生就是高方差的算法
比如KNN算法，一般来说非参数学习都是高方差算法。因为他们不对数据做任何假设。
## 有一些算法天生就是高偏差的算法
比如线性回归算法，一般来说参数学习都是高偏差的。因为他们对数据有很强的假设。

## 机器学习的主要挑战是方差而不是偏差
解决高方差的常用手段
1. 降低模型复杂度
2. 减少数据维度，降噪
3. 增加样本数
4. 使用验证集
5. **模型正则化**

## 模型正则化（Regularization）

限制参数的大小

### 原理
在回归问题中，我们的目标都是使$\sum_{i=1}^{m}{(y^i-\theta_0-\theta_1X_1^i\cdots-\theta_nX_n^i)^2}$尽可能的小,等同于求$J(\theta)=MSE(y,\hat{y};\theta)$尽可能的小

加入模型正则化：
$$J(\theta)=MSE(y,\hat{y};\theta)+\alpha\frac{1}{2}\sum_{i=1}^{n}{\theta_i^2}$$

这样普通线性回归加入了一个L1正则化变成了一个领回归（Ridge Regression）

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
x = np.random.uniform(-3.0, 3.0, size=100)
X = x.reshape(-1, 1)
y = 0.5*x + 3 + np.random.normal(0, 1, size=100)
X.shape
```

```python
plt.scatter(x, y)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split


def PolynomialRegression(degree):
    return Pipeline([
        ("poly",PolynomialFeatures(degree=degree)),
        ("std_scaler",StandardScaler()),
        ("lin_reg",LinearRegression())
    ])
```

```python
np.random.seed(666)

X_train,X_test,y_train,y_test = train_test_split(X,y)
```

```python
from sklearn.metrics import mean_squared_error

poly20_reg = PolynomialRegression(20)
poly20_reg.fit(X_train,y_train)

y_predict = poly20_reg.predict(X_test)

mean_squared_error(y_test, y_predict)
```

```python
X_plot = np.linspace(-3.0, 3.0, 100).reshape(-1,1)
y_plot = poly20_reg.predict(X_plot)

plt.scatter(X, y)
plt.plot(X_plot[:,0],y_plot,color='r')
plt.axis([-3,3,0,6])
```

```python
def plot_model(model):
    X_plot = np.linspace(-3.0,3.0,100).reshape(-1,1)
    y_plot = model.predict(X_plot)

    plt.scatter(X, y)
    plt.plot(X_plot[:,0],y_plot,color='r')
    plt.axis([-3,3,0,6])
    
plot_model(poly20_reg)
```

## 使用岭回归

```python
from sklearn.linear_model import Ridge

def RidgeRegression(degree, alpha):
    return Pipeline([
        ("poly",PolynomialFeatures(degree=degree)),
        ("std_scaler",StandardScaler()),
        ("ridge_reg",Ridge(alpha=alpha))
    ])
```

```python
rid_reg = RidgeRegression(20, 1)
rid_reg.fit(X_train,y_train)
y_predict = rid_reg.predict(X_test)

mean_squared_error(y_test,y_predict)
```

```python
plot_model(rid_reg)
```

```python
rid_reg = RidgeRegression(20, 0.0001)
rid_reg.fit(X_train,y_train)
y_predict = rid_reg.predict(X_test)

mean_squared_error(y_test,y_predict)
```

```python
plot_model(rid_reg)
```

```python
rid_reg = RidgeRegression(20, 100)
rid_reg.fit(X_train,y_train)
y_predict = rid_reg.predict(X_test)

mean_squared_error(y_test,y_predict)
```

```python
plot_model(rid_reg)
```

## 套索回归：一种具有选择功能的L2正则化

公式：
$$J(\theta)=MSE(y,\hat{y};\theta)+\alpha\sum_{i=1}^{n}{|\theta_i|}$$尽可能小

```python
from sklearn.linear_model import Lasso

def LassoRegression(degree, alpha):
    return Pipeline([
        ("poly",PolynomialFeatures(degree=degree)),
        ("std_scaler",StandardScaler()),
        ("lasso_reg",Lasso(alpha=alpha))
    ])
```

```python
las_reg = LassoRegression(20,0.1)
las_reg.fit(X_train,y_train)
y_predict = las_reg.predict(X_test)

mean_squared_error(y_test, y_predict)
```

```python
plot_model(las_reg)
```

LASSO回归趋向于使一部分特征值的theta值变为0，所以他可以用来做特征选择使用
